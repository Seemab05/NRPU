# Split Data into Sentences

```python
# Spliting the Articles into Sentences and storing in a new DataFrame with article IDs
from urduhack.tokenization import sentence_tokenizer

# sub = subset21.reset_index(drop=True)
sub = merged.reset_index(drop=True)

newsids = [] # list to store ids
senlist = [] # list to store separated sentences
s = []
for index, id in enumerate(sub['id']):
  text = sub['News'][index]  # get each article
  if text == 'error':
    newsids.append(id)
    senlist.append('error')
  elif text != 'error':
    sentences = sentence_tokenizer(text)  # split into sentences
  # s.append(len(sentences))
  # if 0 in s:
    # print(f'id: {id}\n news: {text}\n sentences: {sentences}')
    # break
    for i, sentence in enumerate(sentences):
      newsids.append(id)
      senlist.append(sentence)


print(len(senlist))

df2 = pd.DataFrame(columns=['news_id', 'sentences'])  # Create new dataframe

df2['news_id'] = newsids
df2['sentences'] = senlist

df2.shape, df2.head(2)
```

# Get Sentiment Predictions

0 --> neg, 1 --> neu, 2 --> pos

```python
uncleaned = df2['sentences'].tolist()

inputs_unclean = tokenizerSA(uncleaned, padding=True, truncation=True, return_tensors='tf')

logits_unclean=modelSA.predict(dict(inputs_unclean), batch_size=8, verbose=1).logits
prob_unclean=tf.nn.softmax(logits_unclean, axis=1).numpy()
predictions_unclean=np.argmax(prob_unclean, axis=1)
df2['pred_unclean'] = predictions_unclean
df2.head()
```

# Majority Voting

```python
'''
if pos = neg --> neu
if pos = neg = neu --> neu
if pos = neu --> pos
if neg = neu --> neg
'''
from collections import Counter

def find_majority(votes):
    vote_count = Counter(votes)
    top_two = vote_count.most_common(2)
    if len(top_two)>1 and top_two[0][1] == top_two[1][1]:
        if top_two[0][0] != 1 and top_two[1][0] != 1:
          # It is a tie so neutral
          return 1     # 0 --> neg, 1 --> neu, 2 --> pos
        elif top_two[0][0] == 1:
          return top_two[1][0]
        elif top_two[1][0] == 1:
          return top_two[0][0]
    return top_two[0][0]

find_majority([0,0,0,2,2,1,1,1]) # It is a tie
```
